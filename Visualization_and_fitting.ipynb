```
import itertools
import numpy as np
num_time_step_features = 4

samples_control = np.array([list(itertools.chain.from_iterable(
    [list(df_as_control.loc[i+j,]) 
     for j in range(num_time_step_features+1)])) 
           for i in range(df_as_control.shape[0]-(num_time_step_features+1))])

print("Control: ")
print("Shape should be: {} by {}".format(
    df_as_control.shape[0]-(num_time_step_features+1),
    (num_time_step_features+1)*len(allBands)))
print("Shape is: ", samples_control.shape)

samples_music = np.array([list(itertools.chain.from_iterable(
    [list(df_as_music.loc[i+j,]) 
     for j in range(num_time_step_features)])) 
           for i in range(df_as_music.shape[0]-num_time_step_features)])

print("Music: ")
print("Shape should be: {} by {}".format(
    df_as_music.shape[0]-num_time_step_features,
    num_time_step_features*len(allBands)))
print("Shape is: ", samples_music.shape)
```

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

lin_regs = [LinearRegression()]*len(allBands)

X = samples_control[:,:num_time_step_features*len(allBands)]
y = samples_control[:,-len(allBands):]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

for i in range(len(allBands)):
  lin_regs[i].fit(X_train,y_train[:,i])
  
scores = [reg.score(X_test, y_test[:,idx]) for idx, reg in enumerate(lin_regs)]

print("Scores (R^2):")
for band, score in zip(allBands, scores):
  print("{}: {}".format(band, score))
  
print("\nAverage score across bands: ",np.mean(scores))
```

```
from sklearn import svm

svm_regs = [svm.SVR(gamma='auto')]*len(allBands)

for i in range(len(allBands)):
  svm_regs[i].fit(X_train,y_train[:,i])
  
scores = [reg.score(X_test, y_test[:,idx]) for idx, reg in enumerate(svm_regs)]

print("Scores (R^2):")
for band, score in zip(allBands, scores):
  print("{}: {}".format(band, score))
  
print("\nAverage score across bands: ",np.mean(scores))
```

```
svm_predictions = pd.DataFrame(np.array([reg.predict(samples_music) for reg in svm_regs]).T,
                               columns = allBands)

lin_predictions = pd.DataFrame(np.array([reg.predict(samples_music) for reg in lin_regs]).T,
                               columns = allBands)

print(lin_predictions.shape)
print(svm_predictions.shape)
```

```
fig, ax = plt.subplots(nrows=3, sharex=True, figsize=(30, 50))

for band in allBands:
  ax[0].plot(df_as_music.loc[num_time_step_features:, band])
  ax[1].plot(lin_predictions[band])
  ax[2].plot(svm_predictions[band])

ax[0].set_title("Actual EEG Signals with music")
ax[1].set_title("Predicted EEG Signals using linear regression")
ax[2].set_title("Predicted EEG Signals using support vector regression")

plt.show()
```
