```python
def plot_all_bands(df, columns, ax=plt):
    for column in columns:
        ax.plot(np.arange(len(df)), df[column])
    ax.legend()
```

```python
plt.figure(figsize=(20,15))
plot_all_bands(df, allBands)
plt.show()
```
